```
import glob
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
import skimage
import pandas as pd
from google.colab import drive
from os import listdir
from os.path import isfile, join
import os # o módulo precisa ser importado apenas uma vez durante a sessão
from skimage.transform import resize
from skimage.io import imread, imsave


drive.mount('/content/drive')
```

```
path = '/content/drive/My Drive/visao'
```

```
#Redimensionando imagens para não estourar memória do colab na leitura. Trocar o caminho na hora de testar
list_files = os.listdir(path)
for filename in list_files: 
  image = imread(path + '/' + filename)
  image_resized = resize(image, (1280, 720), anti_aliasing=True)
  imsave(path+'/resized/'+filename+'.jpg', image_resized)
```

```
#Lendo as imagens com opencv
mypath= path
onlyfiles = [ f for f in os.listdir(mypath) if isfile(join(mypath,f)) ]
images = np.empty(len(onlyfiles), dtype=object)
for n in range(0, len(onlyfiles)):
  images[n] = cv2.imread( join(mypath,onlyfiles[n]))
```
